```python
android = pd.read_csv("3-Issues/1-IssueAnalysis/android.csv", usecols=[7])
android = android.rename(columns={"category": "Category"})
android["Issues"] = 1
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# create dummy pandas dataframes

pctdistance = 1.3
#create subplot figure with having two side by side plots
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(12,6))
print(axes)
# plot first pandas frame in subplot style
pie1 = android.plot(ax = axes[0], kind="pie", y="Issues", autopct="%1.1f%%", fontsize=20, figsize=(10,7), legend=False, labeldistance=None, pctdistance=pctdistance) 
pie1.set_title("Android Issues", fontsize=20)
pie1.set_ylabel("")
pie1.legend(labels=android.index.values, prop={"size":16}, bbox_to_anchor=(1.45,0.1))

# plot second pandas frame in subplot style
pie2 = spring.plot(ax = axes[1], subplots=False, kind="pie", y="Issues", autopct="%1.1f%%", fontsize=20, figsize=(10,7), legend=False, labeldistance=None, pctdistance=pctdistance)
pie2.set_title("Spring Issues", fontsize=20)
pie2.set_ylabel("")

# plt.tight_layout()

fig.savefig("3-Issues/2-Graphics/issues_category.pdf", bbox_inches="tight")
```

```python
android = pd.read_csv("3-Issues/1-IssueAnalysis/android.csv", usecols=[0, 8, 9, 10])
android = android.rename(columns={"documentation": "Documentation", "source code": "Source Code", "configuration file": "Configuration File"})
```

```python
spring = pd.read_csv("3-Issues/1-IssueAnalysis/spring.csv", usecols=[7, 8, 9])
spring = spring.rename(columns={"documentation": "Documentation", "source code": "Source Code", "configuration file": "Configuration File"})
spring["framework"] = "Spring"
```

```python
android = android.groupby(by="framework", as_index=False).sum()
del android["framework"]
```

```python
spring = spring.groupby(by="framework", as_index=False).sum()
del spring["framework"]
```

```python
android = android.T
android.index.name = "Type"
android["Issues"] = android[0]
del android[0]
```

```python
spring = spring.T
spring.index.name = "Type"
spring["Issues"] = spring[0]
del spring[0]
```

```python
android = android.sort_values("Issues")
```

```python
spring = spring.sort_values("Issues")
```
